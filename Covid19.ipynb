```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("covid19_Confirmed_dataset.csv")
data.head()
data.drop(['Lat','Long'],axis=1, inplace=True)
data.head()
```

```python
data.shape
```

```python
data_aggre = data.groupby("Country/Region").sum()
data_aggre.head()
```

```python
data_aggre.shape
```

```python
data_aggre.loc["Albania"].plot()
data_aggre.loc["India"].plot()
data_aggre.loc["Spain"].plot()
plt.legend()
```

```python
data_aggre.loc["Afghanistan"][:50].plot()
```

```python
data_aggre.loc["Afghanistan"].diff().plot()
```

```python
data_aggre.loc["Afghanistan"].max()
```

```python
data_aggre.loc["India"].max()
```

```python
data_aggre.loc["Spain"].max()
```

```python
countries = list(data_aggre.index)
max_infection_rates=[]
for c in countries:
  max_infection_rates.append(pd.to_numeric(data_aggre.loc[c], errors='coerce').diff().max())
data_aggre["Max_infection_rates"]=max_infection_rates
data_aggre

```

```python
c_data = pd.DataFrame(data_aggre["Max_infection_rates"])
c_data
```
